```python
dfa=df.values
y=(dfa[:,-1]=='Pass').astype(float)
```

```python
def gradient(X_norm,y,theta,alpha,m,n,num_it):
    #gradient descent
    temp=np.zeros((m,1))
    for j in range(0,num_it):
        h=(1/(1+np.exp(-1* X_norm @ theta)))
        
        for i in range(m):
            
            temp[i]=theta[i]-(alpha/n)*(np.matmul(X_norm[:,i].T,(h-y)))
        theta=temp   
    return np.asarray(theta)
def cost(X_norm,y,theta):
    return np.sum(np.log(1+np.exp(y*(-1*X_norm @ theta))))/len(y)
    
    
def logistic(X_train,y_train,X_test):
    #trains with logistic regression outputs theta and predicted y
    X_norm= X_train/X_train.max(axis=0)
    X_norm = np.insert(X_norm, 0, values=1, axis=1)
    X_test=X_test/X_test.max(axis=0)
    X_test=np.insert(X_test, 0, values=1, axis=1)
    n,m=np.shape(X_norm)
    num_it=15000
    alpha=0.1
    min_cost=10000
    for i in range(10):
        print('initialisation ' + str(i)+':')
        theta=np.random.rand(m)[:,np.newaxis]*1e-3
        theta=gradient(X_norm,y_train.reshape(n,-1),theta,alpha,m,n,num_it)
        costL=cost(X_norm,y_train.reshape(n,-1),theta)
        if costL < min_cost :
            min_cost= costL
            thetamin=theta
        print(costL)
    print('mincost is : '+ str(min_cost))
    y=(1/(1+np.exp(-1*X_test@thetamin)))
    return thetamin,y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
```

```python
theta,ynew=logistic(X_train,y_train,X_test)
```

```python
y_pred=1*(ynew>0.5)
```

```python
print("Accuracy = " + str(accuracy_score(y_test,y_pred)))
```

```python
y_pred=pd.Series(y_pred.reshape(-1),name='Predicted')
y_test=pd.Series(y_test,name='Actual')
```
